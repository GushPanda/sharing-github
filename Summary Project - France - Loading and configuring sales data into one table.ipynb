```python
# a function that translates the sale_type column
def translate_sale_type(str):
    if str == 'Adjudication':
        output = 'In legal procedures'
    elif str == 'Echange':
        output = 'Exchange'
    elif str == 'Expropriation':
        output = 'Foreclosure'
    elif str == 'Vente':
        output = 'Sale'
    elif str == 'Vente en l\'état futur d\'achèvement':
        output = 'Future Sale'
    elif str == 'Vente terrain à bâtir':
        output = 'Sale of Terrain'
    return output
```

```python
# applying the function
sales_total['sale_type'] = sales_total['sale_type'].apply(translate_sale_type)
```

```python
# checking the translation
sales_total.groupby('sale_type')['sale_type'].count().to_frame()
```

```python
# a function that translates the property_type column
def translate_property_type(str):
    if str == 'Maison':
        output = 'House'
    elif str == 'Appartement':
        output = 'Appartment'
    elif str == 'Dépendance':
        output = 'Property Dependent'
    elif str == 'Local industriel. commercial ou assimilé':
        output = 'Industrial/Commercial'
    else:
        output = np.nan
    return output
```

```python
# applying the function
sales_total['property_type'] = sales_total['property_type'].apply(translate_property_type)
```

### Transforming and normalizing sale_price into an int value to work with easier

```python
# changing the column to string
sales_total['sale_price'] = sales_total['sale_price'].astype(str)
```

```python
# creating a function to remove the ending of ",00"/',20' etc.
def replace_last(string):
    output = string
    head, _sep, tail = output.rpartition(',')
    output = head
    return output
```
